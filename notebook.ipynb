```python
import os
import anthropic
import google.generativeai as genai
from dotenv import load_dotenv

# 1. Carrega as chaves do seu arquivo .env
load_dotenv()
gemini_key = os.getenv("GEMINI_API_KEY")
claude_key = os.getenv("ANTHROPIC_API_KEY")

print("--- INICIANDO TESTE DE DIAGNÓSTICO DE APIs ---\n")

# ==========================================
# TESTE 1: GOOGLE GEMINI
# ==========================================
print("1. Testando Google Gemini...")
if not gemini_key or gemini_key == "sua_chave_do_google_aqui":
    print("❌ ERRO: Chave do Gemini não encontrada ou inválida no .env")
else:
    try:
        genai.configure(api_key=gemini_key)
        model = genai.GenerativeModel('gemini-1.5-pro-latest')
        response = model.generate_content("Responda apenas com as palavras: 'Gemini Online e Operante'.")
        print(f"✅ SUCESSO GEMINI: {response.text.strip()}")
    except Exception as e:
        print(f"❌ FALHA GEMINI: {e}")

print("\n--------------------------------------------\n")

# ==========================================
# TESTE 2: ANTHROPIC CLAUDE
# ==========================================
print("2. Testando Anthropic Claude...")
if not claude_key or claude_key == "sua_chave_da_anthropic_aqui":
    print("❌ ERRO: Chave do Claude não encontrada ou inválida no .env")
else:
    try:
        client = anthropic.Anthropic(api_key=claude_key)
        message = client.messages.create(
            model="claude-3-5-sonnet-latest",
            max_tokens=20,
            messages=[
                {"role": "user", "content": "Responda apenas com as palavras: 'Claude Online e Operante'."}
            ]
        )
        print(f"✅ SUCESSO CLAUDE: {message.content[0].text.strip()}")
    except anthropic.APIStatusError as e:
        print(f"❌ FALHA CLAUDE (Erro de API): Status {e.status_code} - {e.response.json()}")
    except Exception as e:
        print(f"❌ FALHA CLAUDE (Outro Erro): {e}")

print("\n--- FIM DO DIAGNÓSTICO ---")
```
